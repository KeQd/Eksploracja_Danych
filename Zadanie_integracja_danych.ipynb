```python
########################################
### Zadanie 2

import numpy as np
import pandas as pd
from scipy import stats
from google.colab import files

uploaded = files.upload()
df = pd.read_csv(next(iter(uploaded)))

zakladki = {
    'UE': 'Europe',
    'USA': 'North America (excl. USA)',
    'Chiny': 'China',
    'Indie': 'India'
}

with pd.ExcelWriter('kraje.xlsx') as writer:
    for sheet_name, entity_value in zakladki.items():
        df_kraj = df[df['Entity'] == entity_value]
        df_kraj.to_excel(writer, sheet_name=sheet_name, index=False)

files.download('kraje.xlsx')
```

```python
########################################
### Zadanie 3

import numpy as np
import pandas as pd
import sqlite3
from scipy import stats
from google.colab import files

uploaded = files.upload()
df = pd.read_csv(next(iter(uploaded)))

kraje = ['Poland', 'Germany', 'Spain']
df_filtered = df[df['Entity'].isin(kraje)]

df_final = df_filtered[['Entity', 'Year', 'Annual CO₂ emissions (per capita)']]
df_final = df_final.rename(columns={
    'Entity': 'kraj',
    'Year': 'rok',
    'Annual CO₂ emissions (per capita)': 'emisja_CO2'
})

conn = sqlite3.connect('emisje.db')
df_final.to_sql('emisja_energia', conn, if_exists='replace', index=False)

c = conn.cursor()
c.execute("SELECT * FROM emisja_energia")
rows = c.fetchall()
for row in rows:
    print(row)

conn.close()
```
